## Descriptive Statistics

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'salary_data.csv'
df = pd.read_csv(file_path)

# 1. Descriptive Statistics
print("Descriptive Statistics:")
print(df.describe())

```

## Distribution of the 'Salary' column

```python

plt.figure(figsize=(8, 5))
sns.histplot(df['Salary'], bins=10, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

## Count plot for categorical variables (Job_Role)

```python
plt.figure(figsize=(9, 4))
sns.countplot(x='Job_Role', data=df, palette='Set2')
plt.title('Job Role Distribution')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=40)
plt.show()
```

## Correlation Matrix

## Box plot for Salary by Education level

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Education', y='Salary', data=df, palette='Set3')
plt.title('Salary Distribution by Education Level')
plt.xlabel('Education')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

## Salary comparison by Class Label (Above Median, Below Median)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Class_Label', y='Salary', data=df, palette='Set1')
plt.title('Salary Comparison by Class Label')
plt.xlabel('Class Label')
plt.ylabel('Salary')
plt.xticks([0, 1], ['Below Median', 'Above Median'])
plt.show()
```
